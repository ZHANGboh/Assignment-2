```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://finance.yahoo.com/quote/%5EHSI/history?period1=1477238400&period2=1503244800&interval=1d&filter=history&frequency=1d'
r=requests.get(url)
r.ecoding='utf-8'
html_str=r.text
data=BeautifulSoup(html_str,"html.parser")
a=data.find_all('td')
```

```python
list=[]
for i in range(0,len(a)):
    list.append(a[i].text.strip())
```

```python
import csv
import pandas as pd
```

```python
for i in range(0,len(list)):
    if i%7!=0:
        list[i]=list[i].replace(',','')
list
```

```python
with open('Hang Seng Index.csv','w',newline='')as f:
    writer=csv.writer(f)
    header=['date','open','high','low','close','adj close','volume']
    writer.writerow(header)
    for i in range(0,100):
        writer.writerow([list[i*7],list[i*7+1],list[i*7+2],list[i*7+3],list[i*7+4],list[i*7+5],list[i*7+6]])
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df['datetime'] = df['date'].apply(parse_datetime)
df
```
